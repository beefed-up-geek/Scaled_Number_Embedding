## Qwen/Qwen2.5-Math-1.5B

```python
from transformers import AutoTokenizer, AutoModel

# DeepSeek 7B 모델의 토크나이저 및 모델 불러오기
model_name = "Qwen/Qwen2.5-Math-1.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# 테스트 문장
text = "1+11+111+1111+11111"

# 토크나이징 수행
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text, add_special_tokens=False)

# 임베딩 차원 확인
embedding_dim = model.config.hidden_size  # 모델의 히든 크기 (임베딩 차원)

# 출력
print("토큰화된 결과:", tokens)
print("토큰 ID:", token_ids)
print("디코딩된 텍스트:", tokenizer.decode(token_ids))
print("임베딩 차원:", embedding_dim)
```
